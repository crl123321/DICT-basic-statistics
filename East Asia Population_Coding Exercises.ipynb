```python
import pandas as pd

df = pd.read_csv("population_by_country_2020.csv")

ea = df[df["Country"].isin(["China", "South Korea", "North Korea", "Macau", "Hong Kong", "Tibet", "Taiwan", "Japan", "Mongolia"])]
ea
```

```python
ea.describe()
```

```python
import matplotlib.pyplot as plt

country = ea["Country"]
population = ea["Population (2020)"]

fig = plt.figure(figsize = (16, 10))

plt.barh(country, population)
plt.title("2020 Population by Country (SEA)", loc = "center")

plt.show()
```
